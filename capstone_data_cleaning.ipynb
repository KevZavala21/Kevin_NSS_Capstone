```python
tourism_arrivals = pd.read_csv('../data/tourism_arrivals_bycountry.csv')
tourism_arrivals = tourism_arrivals[['Country Code', 'Country Name', 'Indicator Name',
                                         '2018']].copy()
tourism_arrivals = tourism_arrivals.sort_values(by=['2018'], ascending=False)
tourism_arrivals = tourism_arrivals[~tourism_arrivals['Country Code'].isin(['WLD', 'HIC','OED', 'PST',
                                                                           'ECS', 'IBT', 'LMY', 'IBD',
                                                                           'MIC', 'EUU', 'EMU', 'UMC',
                                                                           'LTE', 'EAS', 'EAR', 'TEC',
                                                                           'EAP', 'TEA', 'LMC', 'ECA',
                                                                           'LCN', 'MEA', 'ARB', 'NAC',
                                                                           'TLA', 'CEB', 'LAC', 'MNA',
                                                                           'TMN', 'SST', 'OSS', 'LDC',
                                                                           'SAS', 'TSA', 'CSS', 'PSS'])]
tourism_arrivals = tourism_arrivals[tourism_arrivals['2018'].notna()]
tourism_arrivals.rename(columns={'Country Code': 'Code', 'Country Name': 'Country'}, inplace=True)
tourism_arrivals
```

```python
tourism_arrivals.to_csv('tourism_arrivals.csv')
```
