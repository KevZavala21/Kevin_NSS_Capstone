```python
tourism_arrivals = pd.read_csv('../data/tourism_arrivals_bycountry.csv')
tourism_arrivals = tourism_arrivals[['Country Name', 'Country Code', 'Indicator Name',
                                         '2018']].copy()
tourism_arrivals = tourism_arrivals.sort_values(by=['2018'], ascending=False)
tourism_arrivals = tourism_arrivals[~tourism_arrivals['Country Code'].isin(['WLD', 'HIC','OED', 'PST',
                                                                           'ECS', 'IBT', 'LMY', 'IBD',
                                                                           'MIC', 'EUU', 'EMU', 'UMC',
                                                                           'LTE', 'EAS', 'EAR', 'TEC',
                                                                           'EAP', 'TEA', 'LMC', 'ECA',
                                                                           'LCN', 'MEA', 'ARB', 'NAC',
                                                                           'TLA', 'CEB', 'LAC', 'MNA',
                                                                           'TMN', 'SST', 'OSS', 'LDC',
                                                                           'SAS', 'TSA', 'CSS', 'PSS'])]
tourism_arrivals = tourism_arrivals[tourism_arrivals['2018'].notna()]
tourism_arrivals
```

```python
country_name = ['France', 'Spain', 'United States', 'China', 'Italy']
tourist_count_millions = [89322000, 82773000, 79745920, 62900000, 61567200]
color = ['blue', 'red', 'gray', 'yellow', 'green']
plt.bar(x = country_name, height = tourist_count_millions, color = color)
plt.show()
```
